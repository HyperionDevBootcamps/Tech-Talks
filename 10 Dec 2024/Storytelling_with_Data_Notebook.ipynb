## Sales by Region

A **bar chart** is used to compare values across different categories.

- The **x-axis** represents the categories, in this case, the **regions** East, West, Central, and South.
- The **y-axis** represents the total **sales** for each region.

By using a bar chart, we can compare sales performance across regions and identify which regions contribute the most or least to overall sales.


```python
# Group data by region and calculate total sales for each subgroup
region_data = data.groupby('Region').sum()[['Sales']]

# Bar chart for sales totals by region
region_data.plot(kind='barh', figsize=(10, 6), color=['lightblue'], edgecolor='black')
plt.title("Total Sales by Region")
plt.xlabel("Amount ($)")
plt.ylabel("Region")
plt.grid() # optional
plt.show() # optional when using notebook, must include if creating python file i
```

### The Story

The West and East regions appear to generate the highest sales, indicating these areas may be the strongest markets or have the most engaged customers. The Central region has moderate sales, suggesting room for improvement. The South region has the lowest sales, signaling possible challenges which need to be investigated.

## Distribution of Sales

A **histogram** is a graphical representation of the distribution of a single continuous variable.

- The **x-axis** represents the range of the variable, in this case, the **sales values**.
- The **y-axis** represents the frequency or count of observations that fall within each range (or bin).

By plotting a histogram of sales, we can see how frequently certain sales values occur, helping us understand the spread and identify any patterns or outliers in the data.


```python
# Plot histogram of sales amounts
plt.figure(figsize=(10, 10))
plt.hist(data['Sales'], bins=100, color='orange', edgecolor='black', alpha=0.7)
plt.title("Distribution of Sales Amounts")
plt.xlabel("Sales ($)")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

### The Story

The majority of transactions have small sales amounts, as shown by the tall bars near the lower range. This suggests that the business relies heavily on a large number of low-value orders. There are a few high-value transactions, as indicated by the extended tail to the right (the bars for the high-value sales are not visible because they are so few). The business might benefit from analysing these cases (do they bulk purchases, premium product sales, special customers, etc.).
